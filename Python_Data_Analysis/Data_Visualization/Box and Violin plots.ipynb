```python
# Standard imports
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

from numpy.random import randn
from scipy import stats
%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
sns.boxplot(data=[data1,data2]) # must specify data kwarg otherwise only 1 plot will show
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
sns.boxplot(data=[data1,data2], orient='h') # use 'orient' keyword isntead of vert. the latter causes an error.
```

```python
# Normal Dist

data1 = stats.norm(0,5).rvs(100)

# Two gamma dist. Concatenated together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                       -1 * stats.gamma(5).rvs(50)])

# Box plot both data1 and data2
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
sns.violinplot(data=[data1,data2])
```

```python
sns.violinplot(data1,bw=0.01)
```

```python
sns.violinplot(data1,inner='stick', orient='v')
```
